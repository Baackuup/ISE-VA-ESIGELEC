## <b>Exercice 5.2 : D√©tection et tracking d'objets par MobileNetV2-SSD-lite</b> 

L'objectif de cet exercice est la <b>d√©tection et tracking d'objets</b> par <b>MobileNetV2-SSD-lite</b> avec les datasets <b>Pascal-VOC</b>

Les diff√©rentes √©tapes sont les suivantes :
- apprentissage (sur GPU) √† partir du backbone pr√©-entra√Æn√© (c'est √† dire apprentissage from scratch de la partie SSD)
- ajout du tracking dans le programme de test sur vid√©os
- test sur au moins 2 vid√©os (au moins 10s chacune)



```python
import torch
import torchvision
from torchvision.models.detection import ssdlite320_mobilenet_v3_large
from torchvision.models.detection.ssd import SSDClassificationHead
import torchvision.transforms as transforms
from torch.utils.data import DataLoader
from torchvision.datasets import VOCDetection

# D√©finition des hyperparam√®tres
batch_size = 16
num_epochs = 20
learning_rate = 1e-3

# Pr√©paration du dataset
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Resize((320, 320)),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

train_dataset = VOCDetection(
    root="./data",
    year="2007",
    image_set="train",
    download=True,
    transform=transform
)
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Chargement du mod√®le MobileNetV2-SSD-lite
model = ssdlite320_mobilenet_v3_large(pretrained=True)
num_classes = 21  # 20 classes et 1 background
model.head.classification_head = SSDClassificationHead(in_channels=576, num_classes=num_classes)
model.to("cuda")

# Optimiseur et fonction de perte
optimizer = torch.optim.Adam(model.parameters(), lr=learning_rate)
criterion = torch.nn.CrossEntropyLoss()
```

```python
# Boucle d'entra√Ænement
for epoch in range(num_epochs):
    model.train()
    total_loss = 0

    for images, targets in train_loader:
        images = [image.to("cuda") for image in images]
        targets = [{k: v.to("cuda") for k, v in t.items()} for t in targets]

        optimizer.zero_grad()
        loss_dict = model(images, targets)
        losses = sum(loss for loss in loss_dict.values())
        losses.backward()
        optimizer.step()

        total_loss += losses.item()

    print(f"Epoch [{epoch+1}/{num_epochs}], Loss: {total_loss/len(train_loader):.4f}")

# Sauvegarde du mod√®le entra√Æn√©
torch.save(model.state_dict(), "mobilenetv2_ssd.pth")
```

```python
from sort import Sort  # T√©l√©charger l'impl√©mentation SORT externe
import cv2

# Initialisation du tracker
tracker = Sort()

# Chargement du mod√®le entra√Æn√©
model.load_state_dict(torch.load("mobilenetv2_ssd.pth"))
model.eval()

# Fonction pour le traitement des vid√©os
def process_video(video_path, output_path):
    cap = cv2.VideoCapture(video_path)
    frame_width = int(cap.get(3))
    frame_height = int(cap.get(4))
    out = cv2.VideoWriter(output_path, cv2.VideoWriter_fourcc(*'XVID'), 30, (frame_width, frame_height))

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        # Pr√©traitement
        image = transforms.ToTensor()(frame).unsqueeze(0).to("cuda")

        # D√©tection des objets
        with torch.no_grad():
            detections = model(image)[0]

        # Formatage des d√©tections pour le tracker
        dets = []
        for box, score, label in zip(detections['boxes'], detections['scores'], detections['labels']):
            if score > 0.5:  # Seuil de confiance
                x1, y1, x2, y2 = box.cpu().numpy()
                dets.append([x1, y1, x2, y2, score.cpu().item()])

        dets = torch.tensor(dets)
        tracks = tracker.update(dets)

        # Affichage des r√©sultats
        for track in tracks:
            x1, y1, x2, y2, track_id = track
            cv2.rectangle(frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
            cv2.putText(frame, f"ID: {int(track_id)}", (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

        out.write(frame)

    cap.release()
    out.release()
    cv2.destroyAllWindows()
```

```python
# Test sur une premi√®re vid√©o
process_video("video1.mp4", "output1.avi")

# Test sur une deuxi√®me vid√©o
process_video("video2.mp4", "output2.avi")
```
